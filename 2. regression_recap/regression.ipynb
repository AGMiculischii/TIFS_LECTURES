```python
from imperio import TargetImputationTransformer
from sklearn.preprocessing import OrdinalEncoder
```

```python
tit = TargetImputationTransformer(index=[12])
```

```python
X = df.drop(["price"], axis=1).values
y = df["price"].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
ordinal_encoder = OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1)
X_12 = ordinal_encoder.fit_transform(X_train[:, 12].reshape(-1, 1))
X_test_12 = ordinal_encoder.transform(X_test[:, 12].reshape(-1, 1))
```

```python
X_12.reshape(1,-1)[0]
```

```python
X_train[:, 12] = X_12.reshape(1,-1)[0]
X_test[:, 12] = X_test_12.reshape(1,-1)[0]
```

```python
X_train
```

```python
X_train_tr = tit.fit_transform(X_train, y_train)
X_test_tr = tit.transform(X_test)
```

```python
X_train_tr
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_tr = scaler.fit_transform(X_train_tr)
X_test_tr = scaler.transform(X_test_tr)
```

```python
model_list = [
    LinearRegression(),
    KNeighborsRegressor(),
    DecisionTreeRegressor()
]
```

```python
for model in model_list:
    model.fit(X_train_tr, y_train/1000000)
    y_pred = model.predict(X_test_tr)
    print(model.__class__)
    print(mean_squared_error(y_pred, y_test/1000000))
```

```python
for model in model_list:
    model.fit(X_train_tr, y_train/1000000)
    y_pred = model.predict(X_test_tr)
    print(model.__class__)
    print(mean_squared_error(y_pred, y_test/1000000))
```
