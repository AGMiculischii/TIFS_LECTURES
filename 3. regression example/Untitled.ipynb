```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("avocado.csv")
```

```python
df
```

```python
df = df.drop(["Unnamed: 0", "Date", "region"], axis=1)
```

```python
df
```

```python
df['type'].value_counts()
```

```python
df['type'] = df['type'].map(
    {
        "conventional" : 0, "organic" : 1
    }
)
```

```python
df
```

```python
X = df.drop(['AveragePrice'], axis=1).values
y = df['AveragePrice'].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
linear = LinearRegression()
linear.fit(X_train, y_train)
y_pred = linear.predict(X_test)
```

```python
mean_squared_error(y_pred, y_test)
```

```python
plt.scatter(y_pred, y_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_tr = scaler.fit_transform(X_train)
X_test_tr = scaler.transform(X_test)
```

```python
linear = LinearRegression()
linear.fit(X_train_tr, y_train)
y_pred_tr = linear.predict(X_test_tr)
```

```python
mean_squared_error(y_pred_tr, y_test)
```
